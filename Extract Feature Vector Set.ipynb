```python
fvs = []
for text in texts:
    
        tmp = []

        # fv1: number of words
        words = text.split()
        tmp.append(len(words))

        #for word in words:
        #fv2: is first word capitalized?
        if words[0][0].isupper():
            tmp.append('true')
        else:
            tmp.append('false')

        #fv3: contains / or -
        found = False
        for word in words:
            if '/' in word or '-' in word:
                found = True
                break
                
        if found:
            tmp.append('true')
        else:
            tmp. append('false')

        #fv4: isDigit?
        found = False
        for word in words:
            if word.isdigit():
                found = True
                break
                
        if found:
            tmp.append('true')
        else:
            tmp.append('false')

        #fv5: are all characters capitalized?
        found = False
        for word in words:
            for c in word:
                if c.isupper():
                    found = True
                    break
                
        if found:
            tmp.append('true')
        else:
            tmp.append('false')

        #fv6: starts with brand name?
        if words[0] in Brand:
            tmp.append('true')
        else:
            tmp.append('false')

        #fv7: is a cateogry word from dictionary?
        found = False
        for word in words:
            if word.lower() in ['laptop','notebook','desktop','tablet']:
                found = True
                break
                
        if found:
            tmp.append('true')
        else:
            tmp.append('false')

        #fv8: is contain ()?
        found = False
        for word in words:
            if '(' in word:
                found = True
                break
        if found:
            tmp.append('true')
        else:
            tmp.append('false')

        #fv9: is contains a string of alphanumeric characters followed by - ?
        found = False
        for word in words:
            if len(re.findall('\w+-\w',word))!=0:
                found = True
                break
        if found:
            tmp.append('true')
        else:
            tmp.append('false')

        #fv10: is a year?
        found = False
        for word in words:
            if '2000'<=len(re.findall('\d+',word))<='2017':
                found = True
                break
        if found:
            tmp.append('true')
        else:
            tmp.append('false')

        #fv11: Inches?
        found = False
        for word in words:
            if word.lower() in ['inch','inches','"']:
                found = True
                break
        if found:
            tmp.append('true')
        else:
            tmp.append('false')

        #fv12: Core?
        found = False
        for word in words:
            if word.lower() is 'core':
                found = True
                break
        if found:
            tmp.append('true')
        else:
            tmp.append('false')

        #fv13: starts with 'i'
        found = False
        for word in words:
            if word.lower().startswith('i'):
                found = True
                break
        if found:
            tmp.append('true')
        else:
            tmp.append('false')

        #fv14: two words concatenated into one string
        found = False
        for word in words:
            if (sum(1 for c in word if c.isupper()))>1:
                found = True
                break
        if found:
            tmp.append('true')
        else:
            tmp.append('false')

        writer.writerow(tmp)
    
        
#fvs.append(tmp)
#len(fvs)
```

## Extracting brand name from two datasets

```python
#healper function for adding brand name in a set
def extractBrandName(brandList,brandName):
    for brand in brandList:
        print brand
        if brand.isalpha():
            brandName.add(brand)
```

```python
# reading amazon dataset
amazon = pd.read_csv('amazon.csv',sep=',', header='infer')
amazon.head()
```

```python
amazonTitle = set()
#extracting brand name from title
brandTokens = amazon.Title.str.split().str.get(0)
extractBrandName(brandTokens,amazonTitle)
amazonTitle
```
