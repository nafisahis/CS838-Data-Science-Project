```python
import os
import re

selected_dir = "Downloads/selected_texts/251-500/"
new_dir = "Downloads/selected_texts/new_251-500/"
selected_files = os.listdir(selected_dir)

f = open("Downloads/product_names.txt", 'r')
names = f.readlines()

new_names = []

for name in names:
    if name != '\n':
        new_names.append(name[:-1])
```

```python
for file in selected_files:
    loc = os.path.join(selected_dir, file)
    new_loc = os.path.join(new_dir, file)

    with open(loc, 'r') as file:
        text = file.read().replace('\n', '')
    
    for name in new_names:
        target = name
        offset = 0
    
        for m in re.finditer(target, text):
            start = m.start() + offset
            end = m.end() + offset
            text = text[:start] + '<p>' + text[start:end] + '</>' + text[end:]
            offset += 6

    f = open(new_loc, 'w')
    f.write(text)
    f.close()
```
