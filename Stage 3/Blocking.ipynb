```python
import py_entitymatching as em
import pandas as pd
import numpy as np
```

```python
datasets_dir = em.get_install_path() + os.sep + 'datasets'
path_A = datasets_dir + os.sep + 'pcmag_stage3.csv'
path_B = datasets_dir + os.sep + 'amazon_stage3.csv'
```

```python
# Read the CSV files
A = em.read_csv_metadata(path_A,low_memory=False) # setting the parameter low_memory to False  to speed up loading.
B = em.read_csv_metadata(path_B,low_memory=False)
```

```python
# Set 'ID as the keys to the input tables
em.set_key(A,'ID')
em.set_key(B,'ID')
```

```python
#Display lengths of the two csv files
print(len(A))
print(len(B))
```

```python
#Down sample the tables
sample_A, sample_B = em.down_sample(A, B, size=5000, y_param=1, show_progress=False)
```

```python
#Print lengths of the sampled tables
print(len(sample_A))
print(len(sample_B))
```

```python
#Get headers of sampled tables
headers_A = list(A.columns)
headers_B = list(B.columns)
print(headers_A)
print(headers_B)
```

```python
#Print head of sample_A
sample_A.head()
```

```python
#Print head of sample_B
sample_B.head()
```

```python
# Instantiate attribute equivalence blocker object
ab = em.AttrEquivalenceBlocker()
# Instantiate overlap blocker object
ob = em.OverlapBlocker()
```

```python
# Specify the tokenization to be 'word' level and set overlap_size to be 2 on Brand field.
C1 = ob.block_tables(sample_A, sample_B, 'Brand', 'Brand', word_level=True, overlap_size=1,allow_missing=True, 
                   l_output_attrs=headers_A, 
                   r_output_attrs=headers_B,
                  show_progress=False)

# Specify the tokenization to be 'word' level and set overlap_size to be 2 on Title field.
C2 = ob.block_candset(C1, 'Title', 'Title', word_level=True, overlap_size=3, allow_missing=True, show_progress=False)

```

```python
# Specify the tokenization to be 'word' level and set overlap_size to be 2 on 'Processor Brand' field.
C3 = ob.block_candset(C2, 'Processor Brand', 'Processor Brand', word_level=True, overlap_size=1, allow_missing=True, show_progress=False)

# Specify the tokenization to be 'word' level and set overlap_size to be 2 on 'Operating System' field.
C4 = ob.block_candset(C3, 'Operating System', 'Operating System', word_level=True, overlap_size=2, allow_missing=True, show_progress=False)
```

```python
#Use block_tables to apply blocking on candidate C4 with RAM parameter attr equivalence.
C5 = ab.block_candset(C4, l_block_attr='RAM', r_block_attr='RAM', allow_missing=True)
```

```python
#Use block_tables to apply blocking on candidate C5 with Screen size parameter attr equivalence.
C6 = ab.block_candset(C5, l_block_attr='Screen Size (inch)', r_block_attr='Screen Size (inch)', allow_missing=True)
```
