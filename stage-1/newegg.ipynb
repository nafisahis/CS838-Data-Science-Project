```python
import urllib2
import re
import time

links = []

# laptops / notebooks
for page_no in range(1,101) :
    url = "https://www.newegg.com/Laptops-Notebooks/SubCategory/ID-32/Page-" + str(page_no) + "?Tid=6740&PageSize=36&order=BESTMATCH"
    website = urllib2.urlopen(url)
    html = website.read().decode('utf8')
    new_links = re.findall('"//www.newegg.com/Product/Product\.aspx\?Item=\w+"', html)
    links = links + new_links
    time.sleep(0.1)

# gaming laptops
for page_no in range(1,43):
    url = "https://www.newegg.com/Gaming-Laptops/SubCategory/ID-3365/Page-" + str(page_no) + "?Tid=167748&PageSize=36&order=BESTMATCH"
    website = urllib.request.urlopen(url)
    html = website.read().decode('utf8')
    new_links = re.findall('"//www.newegg.com/Product/Product\.aspx\?Item=\w+"', html)
    links = links + new_links
    time.sleep(0.1)

# 2-in-1 laptops / Ultrabooks
for page_no in range(1,36):
    url = "https://www.newegg.com/2-in-1-Laptops-Ultrabooks/SubCategory/ID-3090/Page-" + str(page_no) + "?Tid=20039&PageSize=36&order=BESTMATCH"
    website = urllib.request.urlopen(url)
    html = website.read().decode('utf8')
    new_links = re.findall('"//www.newegg.com/Product/Product\.aspx\?Item=\w+"', html)
    links = links + new_links
    time.sleep(0.1)

# Chromebooks
for page_no in range(1,8):
    url = "https://www.newegg.com/Chromebooks/SubCategory/ID-3220/Page-" + str(page_no) + "?Tid=167750&PageSize=36&order=BESTMATCH"
    website = urllib.request.urlopen(url)
    html = website.read().decode('utf8')
    new_links = re.findall('"//www.newegg.com/Product/Product\.aspx\?Item=\w+"', html)
    links = links + new_links
    time.sleep(0.1)
```

```python
import csv
f = open('newegg.csv', 'w')
fields = ['Title', 'Price', 'Currency', 'Sold by Newegg', 'Rating', 'Brand', 'Model', 'Color', 'CPU Type', 'CPU Speed', 'Number of Cores', 'Screen Size', 'Resolution', 'Touchscreen', 'Operating System', 'GPU/VPU', 'Graphic Type', 'SSD', 'HDD', 'Memory', 'Battery', 'Battery Life', 'Dimensions (W x D x H)', 
          'Weight']
writer = csv.writer(f, delimiter = ',')
writer.writerow(fields)
```

```python
from bs4 import BeautifulSoup

for url in unique:
    inst_page = urllib.request.urlopen("https:" + url)
    inst_html = inst_page.read().decode('utf8')

    # parse html
    soup = BeautifulSoup(inst_html, "lxml")
    
    # fetch item name
    title = soup.title.string
    title = title[:-11]

    attribute_list = ['Title']
    value_list = [title]

    # fetch item price
    price_block = soup.find(itemprop='price')
    
    if price_block is None:
        price = ""
    else:
        price = price_block.get('content')

    attribute_list.append('Price')
    value_list.append(price)

    # fetch currency
    currency_block = soup.find(itemprop='priceCurrency')
    
    if currency_block is None:
        currency = ""
    else:
        currency = currency_block.get('content')

    attribute_list.append('Currency')
    value_list.append(currency)
    
    # fetch seller
    seller_block = soup.find('span', class_="grpNote-ship-by")
    
    if seller_block is None:
        sold_by_newegg = "no"
    else:
        seller = seller_block.text
        if seller != "Newegg":
            sold_by_newegg = "no"
        else:
            sold_by_newegg = "yes"
    
    attribute_list.append('Sold by Newegg')
    value_list.append(sold_by_newegg)
    
    # fetch ratings
    rating_block = soup.find('span', class_="print", style="display: none;")
    
    if rating_block is None:
        rating == ""
    else:
        rating = rating_block.text
        if rating == "Rating:0/5":
            rating = ""
        else:
            rating = rating[7:8]
    
    attribute_list.append('Rating')
    value_list.append(rating)
    
    # fetch other specs
    specs = soup.find('div', id="Specs")
    
    if specs is not None:
        dl_list = specs.find_all('dl')

        for dl in dl_list:
            dt = dl.find('dt')
            dd = dl.find('dd')
            attribute_list.append(dt.text)
            value_list.append(dd.text)

        # print(attribute_list)
        # print(value_list) 
    
    row = []

    for field in fields:
        found = False
    
        for index, attribute in enumerate(attribute_list):
            if field == attribute:
                row.append(value_list[index])
                found = True
                break
        
        if not found:
            row.append("")
        
    print(row)

    writer.writerow(row)
    time.sleep(0.01)
```
