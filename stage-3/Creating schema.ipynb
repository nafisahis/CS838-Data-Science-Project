```python
import os, sys
import pandas as pd

amazonFile = os.path.join('..','dataset','amazon.csv')
pcmagFile = os.path.join('..','dataset','pcmag.csv')

amazonData = pd.read_csv(amazonFile, sep=',')
pcmagData = pd.read_csv(pcmagFile, sep=',')

amazonData = amazonData[['Title','Brand Name','Customer Reviews','Price','Processor','Operating System','RAM','Graphics Coprocessor','Hard Drive','Screen Size','Screen Resolution', 
             'Item Weight']]
pcmagData = pcmagData[['Title','Rating','MSRP','Processor Name','Processor Speed','Operating System','RAM','Graphics Card','Storage Capacity (as Tested)','Screen Size','Resolution','Weight']]
amazonData.head()
```

```python
pcmagData.head()
```

```python
brand_p = []
for title in list(pcmagData['Title']):
    tok = str(title).split(' ')
    brand_p.append(tok[0])
pcmagData.insert(1,'Brand',brand_p)
pcmagData.head()
```

```python
wts_a = []
for w in list(amazonData['Item Weight']):
    wts_a.append(str(w).split(' ')[0])
    
wts_p = []
for w in list(pcmagData['Weight']):
    wts_p.append(str(w).split(' ')[0])
del amazonData['Item Weight'] 
```

```python
amazonData.insert(11,'Weight',wts_a)
amazonData.head()
```

```python
pcmagData['Weight'] = wts_p
```

```python
hardDrive_a = []
for h in list(amazonData['Hard Drive']):
    hardDrive_a.append(str(h).split(' ')[0])

amazonData['Hard Drive'] = hardDrive_a
hardDrive_p = []
for h in list(pcmagData['Storage Capacity (as Tested)']):
    hardDrive_p.append(str(h).split(' ')[0])
del pcmagData['Storage Capacity (as Tested)']
pcmagData.insert(8,'Hard Drive',hardDrive_p)
```

```python
screenSize_a = []
for h in list(amazonData['Screen Size']):
    screenSize_a.append(str(h).split(' ')[0])

amazonData['Screen Size'] = screenSize_a

screenSize_p = []
for h in list(pcmagData['Screen Size']):
    screenSize_p.append(str(h).split(' ')[0])

pcmagData['Screen Size'] = screenSize_p
```

```python
screenResolution_a =  []
for s in list(amazonData['Screen Resolution']):
    tok = str(s).split('-')
    if len(tok) == 3:
        screenResolution_a.append(' x '.join([tok[0],tok[2]]))
    else:
        screenResolution_a.append(tok[0])
amazonData['Screen Resolution'] = screenResolution_a
```

```python
pcmagData = pcmagData.rename(columns = {'Resolution':'Screen Resolution'})
pcmagData = pcmagData.rename(columns = {'MSRP':'Price'})
```

```python
ram_a = []
ram_p = list(pcmagData['RAM'])

for p in list(amazonData['RAM']):
    if 'GB' in str(p):
        tok = str(p).split('GB')
        ram_a.append(' GB'.join([tok[0],'']))
    elif 'MB' in str(p):
        tok = str(p).split('MB')
        ram_a.append(' MB'.join([tok[0],'']))
    else:
        ram_a.append(p)
amazonData['RAM'] = ram_a
```

```python
processorName_a =  []
processorSpeed_a =  []

for p in list(amazonData['Processor']):
    if 'GHz' in str(p):
        tok = str(p).split('GHz')
        if len(tok) > 1:
            processorSpeed_a.append(' GHz'.join([tok[0],'']))
            processorName_a.append(tok[1])
        if len(tok) == 1:
            processorName_a.append(tok[0])
            processorSpeed_a.append('')
    elif 'hertz' in str(p):
        tok = str(p).split('hertz')
        if len(tok) > 1:
            processorSpeed_a.append(' GHz'.join([tok[0],'']))
            processorName_a.append(tok[1])
        if len(tok) == 1:
            processorName_a.append(tok[0])
            processorSpeed_a.append('')
    else:
        processorName_a.append('nan')
        processorSpeed_a.append('nan')
amazonData.insert(4,'Processor Name',processorName_a)
amazonData.insert(5,'Processor Speed',processorSpeed_a)
del amazonData['Processor']
```

```python
amazonData = amazonData.rename(columns = {'Customer Reviews':'Rating','Graphics Coprocessor':'Graphics Card'})
```

```python
amazonData = amazonData.rename(columns = {'Brand Name':'Brand'})
amazonData.head()
```
